### Define windows

```python
zone_list = [((220,300),(1280,720),0.7,(253,191,98),0.8),
             ((400,340),(1280,600),0.7,(222,255,98),0.5),
             ((500,370),(1220,530),0.7,(72,240,120),0.8),
             ((600,395),(1100,470),0.7,(150,255,255),0.3)]
             
scales = [[(340,340),(260,260),(210,210)],
          [(210,210),(160,160),(110,110)],
          [(160,160),(110,110),(75,75)],
          [(75,75),(45,45)]]

image = mpimg.imread('test6.jpg')


all_windows = []
for zone, window_list in zip(zone_list,scales):
    zone_windows = []
    zone_color = zone[3]
    thresh = zone[-1]
    overlap = zone[2]
    for window in window_list:
        windows = slide_window(np.copy(image), x_start_stop=[zone[0][0], zone[1][0]], y_start_stop=[zone[0][1], zone[1][1]], 
                               xy_window=window, xy_overlap=(overlap, overlap))
        zone_windows.extend(windows)
    all_windows.append((zone_windows,zone_color,thresh))
    window_img = draw_boxes(image, zone_windows, color=zone[3], thick=3)                    
plt.imshow(window_img)


zones = np.array(all_windows)
zones = np.array(zones[0:4,0])

zone_lens = []

for c, zone in enumerate(zones,1):
    zone_lens.append(len(zone))
    print('Zone '+str(c)+': ' + str(len(zone)))

print('Total windows: '+ str(sum(zone_lens)))
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
import glob
import time
import pickle
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
from skimage.feature import hog
from lesson_functions_udacity import *
from sklearn.cross_validation import train_test_split


#read file names for test images
fnames = glob.glob('test*.jpg')


svc = pickle.load(open( "pickle/classifier.p", "rb" ))
params = pickle.load(open( "pickle/feature_params.p", "rb" ))


X_scaler = params['scaler']
color_space = params['color_space']
spatial_size = params['spatial_size']
hist_bins = params['hist_bins']
orient = params['orient']
pix_per_cell = params['pix_per_cell']
cell_per_block = params['cell_per_block']
hog_channel = params['hog_channel']
          
images = []
times = []
zone_lens = []
box_list = []

for name in fnames:
    image = mpimg.imread(name)
    draw_image = np.copy(image)
    hot_windows = []
    
    t=time.time()
    for zone_windows in all_windows:
        zone_color = zone_windows[1]
        thresh = zone_windows[-1]
        hot_windows.append((search_windows(image, zone_windows[0], svc, X_scaler, color_space=color_space, 
                                spatial_size=spatial_size, hist_bins=hist_bins, 
                                orient=orient, pix_per_cell=pix_per_cell, 
                                cell_per_block=cell_per_block, 
                                hog_channel=hog_channel, spatial_feat=True, 
                                hist_feat=True, hog_feat=True, thresh = thresh), zone_color))               
    t2 = time.time()
    print(round(t2-t, 2), 'Seconds to search '+name)
    
    zones = np.array(hot_windows)
    zones = np.array(zones[0:4,0])

    zone_l = []

    for c, zone in enumerate(zones,1):
        zone_l.append(len(zone))   
    zone_lens.append(zone_l)
    
    box_list.append(hot_windows)
    
    #window_img = np.copy(base_image)
    for zone in hot_windows:
        zone_color = zone[1]
        window_img = draw_boxes(draw_image,zone[0], color=zone_color, thick=3)   
    images.append(window_img)
    
   
#plot results
f, ([ax1, ax2, ax3], [ax4, ax5, ax6]) = plt.subplots(2, 3, figsize=(20,10))
ax1.imshow(images[0])
ax1.set_title('orange:'+str(zone_lens[0][0]) + '  yellow:'+str(zone_lens[0][1])+
              '  green:'+str(zone_lens[0][2])+ '  cian:'+str(zone_lens[0][3]), fontsize=10)
ax2.imshow(images[1])
ax2.set_title('orange:'+str(zone_lens[1][0]) + '  yellow:'+str(zone_lens[1][1])+
              '  green:'+str(zone_lens[1][2])+ '  cian:'+str(zone_lens[1][3]), fontsize=10)
ax3.imshow(images[2])
ax3.set_title('orange:'+str(zone_lens[2][0]) + '  yellow:'+str(zone_lens[2][1])+
              '  green:'+str(zone_lens[2][2])+ '  cian:'+str(zone_lens[2][3]), fontsize=10)
ax4.imshow(images[3])
ax4.set_title('orange:'+str(zone_lens[3][0]) + '  yellow:'+str(zone_lens[3][1])+
              '  green:'+str(zone_lens[3][2])+ '  cian:'+str(zone_lens[3][3]), fontsize=10)
ax5.imshow(images[4])
ax5.set_title('orange:'+str(zone_lens[4][0]) + '  yellow:'+str(zone_lens[4][1])+
              '  green:'+str(zone_lens[4][2])+ '  cian:'+str(zone_lens[4][3]), fontsize=10)
ax6.imshow(images[5])
ax6.set_title('orange:'+str(zone_lens[5][0]) + '  yellow:'+str(zone_lens[5][1])+
              '  green:'+str(zone_lens[5][2])+ '  cian:'+str(zone_lens[5][3]), fontsize=10)
```

### Heatmaps

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import pickle
import cv2
from scipy.ndimage.measurements import label


draw_imgs = []
heatmaps = []

for f,name in enumerate(fnames):
    image = mpimg.imread(name)   
    heat = np.zeros_like(image[:,:,0]).astype(np.float)
    draw_img = None
    heatmap = None
    
    for z in range(4):
        # Add heat to each box in box list
        heat = add_heat(heat,box_list[f][z][0])

    # Apply threshold to help remove false positives
    heat = apply_threshold(heat,5)

    # Visualize the heatmap when displaying    
    heatmap = np.clip(heat, 0, 255)
    heatmaps.append(heatmap)
    
    # Find final boxes from heatmap using label function
    labels = label(heatmap)
    draw_img = draw_labeled_bboxes(np.copy(image), labels)
    
    draw_imgs.append(draw_img)
    
        
f, (ax1,ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.imshow(draw_imgs[5])
ax1.set_title('Car Positions')
ax2.imshow(heatmaps[5], cmap='hot')
ax2.set_title('orange:'+str(zone_lens[0][0]) + '  yellow:'+str(zone_lens[0][1])+
              '  green:'+str(zone_lens[0][2])+ '  cian:'+str(zone_lens[0][3]), fontsize=10)
    

```
